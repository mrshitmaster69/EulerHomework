Initial Steps:
* Import tools
* Establish initial variables

Inside loop:
1. update theta
2. update theta prime
3. update theta double prime
4. append theta
5. append theta prime
6. append theta double prime

After loop:
* return variables

```python
#import packages
import math
import numpy as np
import matplotlib.pyplot as plt

#Euler method function
def euler(omega_0, theta_0, angvel_0, t_0, t_f, n_steps):
  dt = (t_f - t_0) / n_steps
  theta = [theta_0]
  theta_prime = [angvel_0]
  t = [t_0]
  for i in range(n_steps):
    theta_new = theta[-1] + theta_prime[-1] * dt
    theta_prime_new = theta_prime[-1] - omega_0**2 * math.sin(theta_new) * dt
    t_new = t[-1] + dt
    theta.append(theta_new)
    theta_prime.append(theta_prime_new)
    t.append(t_new)
  return theta, theta_prime, t
```

```python
# Code cell for theta_0 = pi, ang_vel_0 = 0

#run function
theta, thetaprime, t = euler(1, math.pi, 0, 0, 100, 1000)

# plot with matplotlib
plt.figure(figsize=(10,3))
plt.plot(t, theta, label='euler theta')
plt.xlabel('t')
plt.ylabel('theta')
plt.legend()
plt.show()

```

```python
#Code cell for theta_0 = 0, ang_vel_0 = 2

#run function
theta2, thetaprime2, t2 = euler(1, 0, 2, 0, 100, 1000)

#plot function
plt.figure(figsize=(10,3))
plt.plot(t2, theta2, label='euler theta')
plt.xlabel('t2')
plt.ylabel('theta2')
plt.legend()
plt.show()

#plot derivative
plt.figure(figsize=(10,3))
plt.plot(t2, thetaprime2, label='euler angular velocity')
plt.xlabel('t2')
plt.ylabel('theta2')
plt.legend()
plt.show()

```
